```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
z = np.arange(-7, 7, 0.1)
```

```python
phiz = 1.0 / (1.0 + np.exp(-z))
```

```python
plt.xkcd()  # cartoni
```

```python
plt.plot(z, phiz, color='r')

plt.axvline(0, ls='dotted', color='b')
plt.axhline(y=0.5, ls='dotted', color='b')
plt.yticks([0, 0.5, 1])
plt.xlabel('z')
plt.ylabel('$\phi (z)$')
```
